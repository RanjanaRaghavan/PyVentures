# Tic Tac Toe ‚ùå üíö‚≠ï
Welcome to Tic Tac Toe! This fun-filled game lets you play with your choice of X ‚ùå or O ‚≠ï. The goal is to get three of your symbols in a row, either horizontally, vertically, or diagonally. The player who does that first wins the game! Ready to test your luck against the computer? Let‚Äôs start the game!

Description:

This script implements a basic Tic Tac Toe game where:


1.   The game board is represented by a dictionary where keys are cell numbers ('1' to '9') and values are the cell contents (either 'X', 'O', or ' ' for empty).

2.   The player chooses to play as either 'X' or 'O', and the computer picks the opposite.

3.   Players take turns entering their moves (player first, followed by the computer's random move).

4.   The script checks after every move to see if a player or the computer has won, or if the game ends in a draw.

5.   You can challenge yourself to play multiple rounds, and see how quickly you can beat the computer!

```python
#Tic Tac Toe

'''
#Multiline Comment
Our Bopard positions Looks like this

1 | 2 | 3
---------
4 | 5 | 6
---------
7 | 8 | 9

'''
from random import randint

# Function to print Board

def printBoard(board):
  print(board['1'], ' | ' , board['2'] , ' | ', board['3'])
  print('-------------')
  print(board['4'], ' | ' , board['5'] , ' | ', board['6'])
  print('-------------')
  print(board['7'], ' | ' , board['8'] , ' | ', board['9'])

# Play Game Function

def checkWin(board):
  #Check rows
  #first row starts in 1, second in 4 and third is 7
  for row in range(1,8,3):
    if board[str(row)] == board[str(row+1)] == board[str(row+2)] and board[str(row)] != ' ':
      return board[str(row)]

  #Check Colums
  for col in range(1,4):
    if board[str(col)] == board[str(col+3)] == board[str(col+6)] and board[str(col)] != ' ':
      return board[str(col)]

  # Check diagonals
  if (board['1'] == board['5'] == board['9'] and board['1'] != ' ') or \
     (board['3'] == board['5'] == board['7'] and board['3'] != ' '):
    return board['5']

  return None


def playGame(choice,name):

  import random
  #Start with clean board
  board = {
              '1' : ' ',
              '2' : ' ',
              '3' : ' ',
              '4' : ' ',
              '5' : ' ',
              '6' : ' ',
              '7' : ' ',
              '8' : ' ',
              '9' : ' '
          }

  #Assign players their symbol
  player = choice
  if choice.upper() == 'X':
    computer = 'O'
  else:
    computer = 'X'

 #Assume Player always plays first
  count = 0 #To count no of plays
  filledCells = [] # To keep track of already filled Cells
  while count < 9 and checkWin(board) is None:

    print('Choose the cell you want to play',name)
    cell = input()
    board[str(cell)] = choice
    filledCells.append(int(cell))
    printBoard(board)

    if count >= 5 and checkWin(board):
      break

    if count < 9:
    #Computer plays a random int
      compCell = -1
      while compCell ==-1 or compCell in filledCells:
        compCell = random.randint(1,9)

      print('Computer plays',compCell)
      board[str(compCell)] = computer
      filledCells.append(compCell)
      printBoard(board)
      count = count+1

  if checkWin(board) == choice:
      print('Player',name, 'wins')
  elif checkWin(board) == computer:
      print('The Computer Wins')
  else:
      print('Draw the match')

```

```python
#Win - column check
print('Lets play Tic Tac Toeee!!!!')
print('Hi Player, What is your name?')
name = input()
print('Well hello there',name,'Do you want to play as X or O today?')
choice = input()
print('Lets Begin!!!')
playGame(choice,name)
```
